### Categorización y One-Hot Encoding

Carga el dataset preprocesado exportado desde el notebook de EDA.

```python
import pandas as pd
import numpy as np
from collections import Counter

data = pd.read_csv("data_procesado.csv")
print(f"Dataset cargado: {data.shape[0]:,} filas, {data.shape[1]} columnas")
data.head()
```

### Categorizacion de reviews


```python
data
```

```python
# Tasa de reseñas positivas
data['total_reviews'] = data['Positive'] + data['Negative']
data['review_rate'] = data['Positive'] / data['total_reviews'].replace(0, pd.NA)

#data[['Name','total_reviews','review_rate']].head(10)
#data.groupby('total_reviews')[['Name','total_reviews','review_rate']].head(10)
data.sort_values(by='total_reviews',ascending=False)[['Name','total_reviews','review_rate']].head(10)
```

```python
def steam_rating_category(row):
    total = row['total_reviews']
    rate = row['review_rate']
    if pd.isna(rate) or total == 0:
        return 'No reviews'
    if total < 10:
        return 'Too few reviews'
    if rate >= 0.95:
        return 'Overwhelmingly Positive' if total >= 500 else 'Very Positive'
    if rate >= 0.80:
        return 'Very Positive' if total >= 50 else 'Mostly Positive'
    if rate >= 0.70:
        return 'Mostly Positive'
    if rate >= 0.40:
        return 'Mixed'
    if rate >= 0.20:
        return 'Mostly Negative'
    if total >= 500:
        return 'Overwhelmingly Negative'
    if total >= 50:
        return 'Very Negative'
    return 'Negative'

data['review_score_cat'] = data.apply(steam_rating_category, axis=1)

orden = ['Overwhelmingly Positive', 'Very Positive', 'Mostly Positive', 'Mixed',
         'Mostly Negative', 'Very Negative', 'Overwhelmingly Negative', 'Negative',
         'Too few reviews', 'No reviews']
print(data['review_score_cat'].value_counts().reindex(orden, fill_value=0))
```

### Categorización de ventas (Estimated Owners)

```python
def categorize_owners(owners_str):
    if pd.isna(owners_str):
        return "No data"
    lower = int(owners_str.split(" - ")[0].replace(",", "").strip())
    if lower < 20_000:
        return "Niche"
    if lower < 100_000:
        return "Small Indie"
    if lower < 500_000:
        return "Growing"
    if lower < 1_000_000:
        return "Popular"
    if lower < 2_000_000:
        return "Very Popular"
    if lower < 10_000_000:
        return "Top Game"
    if lower < 50_000_000:
        return "Blockbuster"
    return "Mega Hit"

data["sales_cat"] = data["Estimated owners"].apply(categorize_owners)

sales_order = ["Mega Hit", "Blockbuster", "Top Game", "Very Popular",
               "Popular", "Growing", "Small Indie", "Niche", "No data"]
print(data["sales_cat"].value_counts().reindex(sales_order, fill_value=0))
```

### Categorización de desempeño (Peak CCU y Metacritic)

```python
# Peak CCU por percentiles (excluyendo ceros)
ccu_nonzero = data[data["Peak CCU"] > 0]["Peak CCU"]
p25, p50, p75, p90 = ccu_nonzero.quantile([0.25, 0.50, 0.75, 0.90])
print(f"Percentiles CCU (sin ceros): p25={p25:.0f}, p50={p50:.0f}, p75={p75:.0f}, p90={p90:.0f}")

def categorize_ccu(ccu):
    if ccu == 0:
        return "No activity"
    if ccu <= p25:
        return "Very Low"
    if ccu <= p50:
        return "Low"
    if ccu <= p75:
        return "Medium"
    if ccu <= p90:
        return "High"
    return "Very High"

data["ccu_cat"] = data["Peak CCU"].apply(categorize_ccu)
print("Peak CCU category distribution:")
print(data["ccu_cat"].value_counts())

# Metacritic por escala estandar
def categorize_metacritic(score):
    if score == 0:
        return "Not rated"
    if score < 50:
        return "Poor"
    if score < 60:
        return "Mediocre"
    if score < 70:
        return "Average"
    if score < 80:
        return "Good"
    if score < 90:
        return "Very Good"
    return "Excellent"

data["metacritic_cat"] = data["Metacritic score"].apply(categorize_metacritic)
print("Metacritic category distribution:")
metacritic_order = ["Excellent", "Very Good", "Good", "Average", "Mediocre", "Poor", "Not rated"]
print(data["metacritic_cat"].value_counts().reindex(metacritic_order, fill_value=0))
```

### Datos seccionados - Juegos de interés

juegos que cumplen al menos una de las condiciones de éxito:
- **Muy buenas reviews**: `review_score_cat` ∈ `[Overwhelmingly Positive, Very Positive]`
- **Muchas ventas**: `sales_cat` ∈ `[Top Game, Blockbuster, Mega Hit]`

```python
good_reviews = data["review_score_cat"].isin(["Overwhelmingly Positive", "Very Positive"])
good_sales   = data["sales_cat"].isin(["Top Game", "Blockbuster", "Mega Hit"])

data_top = data[good_reviews | good_sales].copy()

print(f"Dataset completo:   {data.shape[0]:,} juegos")
print(f"Dataset filtrado:   {data_top.shape[0]:,} juegos ({data_top.shape[0]/data.shape[0]*100:.1f}%)")
print()
print("Distribucion review_score_cat en el subset:")
print(data_top["review_score_cat"].value_counts())
print()
print("Distribucion sales_cat en el subset:")
print(data_top["sales_cat"].value_counts())
```

```python
# Estadisticas descriptivas del subset vs dataset completo
cols_compare = ["Price", "Peak CCU", "DLC count", "Metacritic score",
                "Average playtime forever", "Positive", "Negative"]

print("=== Dataset completo ===")
print(data[cols_compare].describe().round(2))
print()
print("=== Subset (top games) ===")
print(data_top[cols_compare].describe().round(2))
```

### One-Hot Encoding - por generos y Tags

- **Generos**: OHE completo sobre los 33 géneros únicos.
- **Tags**: Se conservan solo las etiquetas con ≥ 5 000 apariciones en el dataset (59 de 452) para evitar columnas demasiado dispersas que perjudiquen los algoritmos de asociación.

```python
from collections import Counter

# --- Géneros: OHE completo ---
genres_ohe = data["Genres"].fillna("").str.get_dummies(sep=",")
genres_ohe.columns = ["genre_" + c.strip() for c in genres_ohe.columns]

# --- Tags: filtrar por frecuencia mínima ---
tag_counts = Counter()
for val in data["Tags"].dropna():
    for t in val.split(","):
        tag_counts[t.strip()] += 1

frequent_tags = {t for t, c in tag_counts.items() if c >= 5000}

tags_filtered = data["Tags"].fillna("").apply(
    lambda x: ",".join(t.strip() for t in x.split(",") if t.strip() in frequent_tags)
)
tags_ohe = tags_filtered.str.get_dummies(sep=",")
tags_ohe.columns = ["tag_" + c for c in tags_ohe.columns]

# --- Unir al dataframe principal ---
data = pd.concat([data, genres_ohe, tags_ohe], axis=1)

print(f"Columnas de género (genre_*):  {genres_ohe.shape[1]}")
print(f"Columnas de tags  (tag_*):     {tags_ohe.shape[1]}")
print(f"Total columnas binarias nuevas: {genres_ohe.shape[1] + tags_ohe.shape[1]}")
print(f"Shape: {data.shape}")
print()
print("Columnas de género:")
print([c for c in data.columns if c.startswith("genre_")])
print()
print("Tags frecuentes incluidas:")
print([c.replace("tag_","") for c in data.columns if c.startswith("tag_")])
```
